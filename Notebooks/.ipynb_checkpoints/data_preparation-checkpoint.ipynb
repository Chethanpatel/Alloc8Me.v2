```python
from pathlib import Path
import pandas as pd
import os
```

```python
# Get the current working directory (which should be the Notebooks folder)
current_directory = Path(os.getcwd())

# Define the project root directory by navigating up one level from the Notebooks folder
project_root = current_directory.parent

# Construct the path to the CSV file in the dataset folder
cutoff_file_path = project_root / 'dataset' / 'cutoff.csv'
```

```python
df = pd.read_csv(cutoff_file_path)
```

```python
df
```

```python
# List of columns to exclude
columns_to_exclude = ['InstituteId', 'InstituteName', 'BranchCode']

# Create a variable with columns excluded
cut_off_data_columns = [col for col in df.columns if col not in columns_to_exclude]
```

```python
cut_off_data_columns
```

```python
df
```

```python
df.dtypes
```

```python
# Convert float64 columns to integers and replace NaN with empty strings
for column in df.select_dtypes(include='float64').columns:
    df[column] = df[column].fillna(0).astype(int)
```

```python
df
```

```python
df.shape
```

```python
# Construct the path to the CSV file in the dataset folder
cleaned_cutoff_file_path = project_root / 'dataset' / 'cleaned_cutoff.csv'

df.to_csv(cleaned_cutoff_file_path)
```

```python
course_data_file_path = project_root / 'dataset' / 'courses.csv'
course_df = pd.read_csv(course_data_file_path)
```

```python
course_df
```

```python
# Perform a left join between df and courses_df on 'BranchCode' and 'Code'
merged_df = pd.merge(df, course_df, how='left', left_on='BranchCode', right_on='Code')

# Rename the 'course name' column to 'branchname'
merged_df.rename(columns={'course name': 'branchname'}, inplace=True)

# Drop the duplicated 'Code' column
merged_df.drop('Code', axis=1, inplace=True)
```

```python
df2  = merged_df
```

```python
df2
```

```python
college_data_file_path = project_root / 'dynamic_data' / 'college_data.csv'
df2[[ 'InstituteId', 'InstituteName']].drop_duplicates(subset='InstituteId', keep='first')
df2.reset_index(drop=True, inplace=True)
```
