```python
df11 = pd.read_csv( project_root / 'assets' / 'dataset.csv')
```

```python
df11
```

```python
# Define mappings from columns to categories
category_mappings = {
    'Category-1': ['1G', '1K', '1R'],
    'Category-2A': ['2AG', '2AK', '2AR'],
    'Category-2B': ['2BG', '2BK', '2BR'],
    'Category-3A': ['3AG', '3AK', '3AR'],
    'Category-3B': ['3BG', '3BK', '3BR'],
    'Category-GM': ['GM', 'GMK', 'GMR'],
    'Category-SC': ['SCG', 'SCK', 'SCR'],
    'Category-ST': ['STG', 'STK', 'STR']
}

# Initialize an empty DataFrame to store results
result_df = df11.copy()

# Calculate minimum and maximum cutoff rank for each category
for category, columns in category_mappings.items():
    # Replace zeros with NaN to ignore them in min/max calculation
    replaced_df = df11[columns].replace(0, pd.NA)
    
    # Calculate the minimum and maximum values for each row in the category
    min_values = replaced_df.min(axis=1)
    max_values = replaced_df.max(axis=1)
    
    # Replace NaN with 0 (if all values in row were zero)
    min_values = min_values.fillna(0)
    max_values = max_values.fillna(0)
    
    # Assign the minimum and maximum values to the result dataframe
    result_df[f'{category}_min'] = min_values
    result_df[f'{category}_max'] = max_values
```

```python
import pandas as pd
import json

# Load your CSV file
df = pd.read_csv('branch.csv')

# Create a dictionary to store the data in the desired format
institutions = {}

# Iterate through the DataFrame rows and populate the dictionary
for index, row in df.iterrows():
    institution = row['Institution'].strip()
    course = row['Course'].strip()
    
    if institution not in institutions:
        institutions[institution] = []
        
    institutions[institution].append(course)

# Save the dictionary as a JSON file
with open('institutions_courses.json', 'w') as json_file:
    json.dump(institutions, json_file, indent=4)

print("JSON file created successfully.")
```
